```
import spacy
```

```
nlp = spacy.load("en_core_web_lg")
```

```
import numpy as np
```

```
sentencizer = nlp.create_pipe("sentencizer")
nlp.add_pipe(sentencizer)
```

```
doc = nlp(article_text)
doc
```

```
len(list(doc.sents))
```

```
sim_matrix = np.zeros([len(list(doc.sents)),len(list(doc.sents))])
sim_matrix.shape
```

```
doc.sents
```

```
sim_matrix = np.zeros([len(list(doc.sents)),len(list(doc.sents))])
for col in range(len(list(doc.sents))):
  for row in range(len(list(doc.sents))):
    if row != col:
        sim_matrix[col][row] = list(doc.sents)[col].similarity(list(doc.sents)[row])
  
```

```
sim_matrix()
```

```
sim_matrix.shape
```

```
import networkx as nx
```

```
nx_graph = nx.from_numpy_array(sim_matrix)
scores = nx.pagerank(nx_graph)
```

```
nx_graph, scores
```

```
ranked_sentences = sorted(((scores[i],s) for i,s in enumerate(list(doc.sents))), reverse=True)
```

```
ranked_sentences[0][1]
```

```
for i in range(10):
  print(ranked_sentences[i][1])
```
